```python
#Räkna totalt antal bilar lager per bilmärke och skriv ut detta.
```

```python

grouped_märke = df_bil.groupby('company')
```

```python
grouped_märke.count()
#grouped_märke.size()
```

```python
grouped_märke.size()
```

```python
#Skriv ut data om varje bilmärkes dyrast bil.
#1.skrive bara ut information om dyrast bil som är första av märket
```

```python
def getpriceTopn(df,topn):
    sort_df = df.sort_values(by='price',na_position='first')[-topn:]
    return sort_df
grouped_märke.apply(getpriceTopn,topn =1)
```

```python
#2.eller(men bara max price information)
```

```python
print(grouped_märke.price.max())
```

```python
#3. eller (alla bilar information som har dyrast price)
```

```python
dyrast_df = grouped_märke.apply(lambda t : t[t.price == t.price.max()])
dyrast_df=dyrast_df.rename(columns={'company':'car_brands'})
dyrast_df
```
